```python
def create_df(responses: [(int, requests.Response)]) -> pd.DataFrame:
    responses_df = pd.DataFrame([response[1].json()[str(response[0])]["data"] for response in responses])
    responses_df["required_age"] = responses_df["required_age"].astype(int)
    return responses_df

def create_json(responses: [(int, requests.Response)]):
    return {appid: response.json()[str(appid)]["data"] for appid, response in responses}
```
